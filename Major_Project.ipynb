#FBProphet Time Series Analysis

```
import fbprophet
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
data=pd.read_csv('fbp.csv')
data
dicto={'Date':data['Date'],'Active':data['Active']}
df=pd.DataFrame(dicto)
df
df['Active']=df['Active']-df['Active'].shift(1)
df
df.plot()
df_new = df.rename(columns={'Date':'ds','Active':'y'})
df_new=df_new[['ds','y']]
df=df_new
df
model=fbprophet.Prophet()
model.fit(df)
mfd=model.make_future_dataframe(periods=120)
pred=model.predict(mfd)
pred[['ds','yhat','yhat_lower','yhat_upper']].tail(50)
model.plot(pred,xlabel='Date',ylabel='Active cases')
plt.title('Active Cases Forecasting')
```
